```python
import requests
from bs4 import BeautifulSoup as BS
from pathlib import Path

response = requests.get("https://www.ptt.cc/bbs/Gossiping/index.html",cookies = {"over18":"1"})
#response.encoding = "utf-8"
bs = BS(response.text,features = "html.parser")

url = "https://www.ptt.cc"

data_list = bs.find_all("div",class_ = "title")
output = ""
for data in data_list:
    if data.find("a"):
        output += data.find("a").text + "\n"
        output += url + data.find("a").get("href") + "\n\n"

cur_path = Path(".")
new_path = cur_path.joinpath("download","data.txt")
#print(new_path)

file = open(new_path,"w")
file.write(output)
file.close()
```

```python
from pathlib import Path
cur_path = Path(".")
new_path = cur_path.joinpath("download")
num = 0
for img in imgs:
    jpg = requests.get(img.get("src"))
    data = open(f"{new_path}/testagain_{num}.png","wb")
    data.write(jpg.content)
    data.close()
    num += 1
```
